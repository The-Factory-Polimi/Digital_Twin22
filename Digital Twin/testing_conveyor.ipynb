```python
import simpy

def machine(env, queue_out):
    while True:
        # Process time
        print(f"t: {env.now}: Machine Started")
        yield env.timeout(1)

        # Put in the queue
        print(f"t: {env.now}: Machine sent part {env.now}")
        yield queue_out.put(env.now)
        

class Queue():
    def __init__(self, env, name):
        self.env = env
        self.name = name
        self.conveyor = simpy.Store(env)
        self.queue = simpy.Store(env)

    def put(self, part):
        self.conveyor.put(part)
        print(f"t: {self.env.now} - Conveyor received part {part}")
        yield self.env.timeout(2)
        print(f"t: {self.env.now} - Conveyor finish the transportation of {part}")

env = simpy.Environment()
queue = Queue(env, "queue 1")
env.process(machine(env, queue))
env.run(until=20)

```
