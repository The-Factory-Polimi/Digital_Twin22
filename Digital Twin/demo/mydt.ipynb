## My first DT

Importing the library

````python
from dtwinpy.Digital_Twin import Digital_Twin
````

Create your first DT

````python
mydt = Digital_Twin(name= '5s_determ')
````

### Simple Simulations

Running a simple simulation to finish 5 parts

````python
mydt.run_digital_model(maxparts=5, plot= False)
````

Running for a targeted part id to finish the simulation

````python
mydt.run_digital_model(targeted_part_id= 2)
````

Running the simulation until a targeted part id finish a certain cluster

````python
mydt.run_digital_model(targeted_part_id=2, targeted_cluster= 3)
````

### Run Synchronization (based on the last point of the simulation)

Run a simulation until a certain point

````python
mydt = Digital_Twin(name= "5s_determ", copied_realDB= True)
````

````python
mydt.run_digital_model(targeted_part_id= 2, targeted_cluster= 4)
````

Run Synchronization by copying the digital database created by the last simulation

````python
mydt.run_sync(copied_realDB= True)
````
