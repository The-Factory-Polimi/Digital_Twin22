## First Application of Database to the project

Event log strucutre

Timestamp: Time when the acitivity occured

- Resource: Machine ID
- Activity Type: 
    - "s": Started
    - "f": Finished
- Part ID: Part ID involved in the activity 

Creating and Populating the database

```python
import sqlite3


#--- Cleaning table
with sqlite3.connect("databases/digital_model_DB2.db") as digital_model_DB2:

    digital_model_DB2.execute("DROP TABLE IF EXISTS digital_log2")
    digital_model_DB2.commit()


table_name = "digital_log2"
#--- Creating table
with sqlite3.connect("databases/digital_model_DB3.db") as digital_model_DB2:

    digital_model_DB2.execute(f"""
    CREATE TABLE IF NOT EXISTS {table_name} (
        event_id INTEGER PRIMARY KEY,
        timestamp INTEGER,
        machine_id INTEGER,
        activity_type TEXT,
        part_id INTEGER
    )
    """)
    digital_model_DB2.commit()

#--- Writing into the table
with sqlite3.connect("databases/digital_model_DB2.db") as digital_model_DB2:
    for i in range(100):
        timestamp = "Time: " + str(i)
        part_id = "Part " + str(i)
        if i % 2 == 0: 
            machine_id = "Machine " + str(i + 1) 
            activity_type = "Finished"
        else: 
            machine_id = "Machine " + str(i + 2)
            activity_type = "Started"
        

        digital_model_DB2.execute(f"""
        INSERT INTO {table_name} (timestamp, machine_id, activity_type, part_id)
        VALUES (?, ?, ?, ?)""", (timestamp, machine_id, activity_type, part_id))

    digital_model_DB2.commit()
```

Reading the Database

```python
#--- Reading and Printing data from the table
with sqlite3.connect("databases/digital_model_DB2.db") as digital_model_DB2:
    event_points = digital_model_DB2.execute("SELECT * FROM digital_log2")
    for event_point in event_points:
        print(event_point)
```
